##Ubah Data kembali menjadi categorical untuk mempermudah proses analisis

1. DF day

```python
#ubah dataframe day
day_df["yr"] = day_df["yr"].map({
    0:'2011', 1:'2012'
})
day_df["workingday"] = day_df["workingday"].map({
    0:'Holiday', 1:'workingday'
})
day_df["season"] = day_df["season"].map({
    1:'springer', 2:'summer', 3:'fall', 4:'winter'
})
day_df["mnth"] = day_df["mnth"].map({
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
})
day_df["weekday"] = day_df["weekday"].map({
    0:'Sun', 1:'Mon', 2:'Tue', 3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat'
})
day_df["weathersit"] = day_df["weathersit"].map({
    1:'Clear, cloudy', 2:'Mist + Cloudy',
    3:'Light, Snow, Rain, Thunderstorm', 4:'Heavy Rain, Ice, Thunderstorm'
})
day_df.head()
```

2. DF hour

```python
#ubah dataframe hour
hour_df["yr"] = hour_df["yr"].map({
    0:'2011', 1:'2012'
})
hour_df["workingday"] = hour_df["workingday"].map({
    0:'Holiday', 1:'workingday'
})
hour_df["season"] = hour_df["season"].map({
    1:'springer', 2:'summer', 3:'fall', 4:'winter'
})
hour_df["mnth"] = hour_df["mnth"].map({
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
})
hour_df["weekday"] = hour_df["weekday"].map({
    0:'Sun', 1:'Mon', 2:'Tue', 3:'Wed', 4:'Thu', 5:'Fri', 6:'Sat'
})
hour_df["weathersit"] = hour_df["weathersit"].map({
    1:'Clear, cloudy', 2:'Mist + Cloudy',
    3:'Light, Snow, Rain, Thunderstorm', 4:'Heavy Rain, Ice, Thunderstorm'
})
hour_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:Apakah ada pengaruh jumlah permintaan sepeda antara weekday dan holyday?

```python
#visualisasi pertanyaan 1
plt.figure(figsize=(5,5))
sns.barplot(
    x='workingday',
    y='cnt',
    data=day_df,
    palette='viridis')

plt.title('Jumlah Pengguna Sepeda berdasarkan harikerja dan harilibur')
plt.xlabel('')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh cuaca terhadap permintaan sepeda?

```python
#visualisasi pertanyaan 2
plt.figure(figsize=(5,5))
sns.barplot(
    y='weathersit',
    x='cnt',
    data=hour_df,
    palette='viridis')

plt.title('Jumlah Pengguna Sepeda berdasarkan cuaca')
plt.xlabel('Jumlah')
plt.ylabel('Kondisi Cuaca')
plt.show()
```

```python
#visualisasi pertanyaan 2
column = day_df[['temp', 'atemp', 'cnt', 'hum', 'windspeed']]
correlation = column.corr()
plt.figure(figsize=(5, 3))
sns.heatmap(correlation, annot=True, center=0)
plt.title('Korelasi temp, atemp,hum,windspeed dan cnt')
plt.show()
```

###Pertanyaan 3 : Bagaimana penggunaan sepeda berbeda antara pengguna terdaftar dan pengguna tidak terdaftar?

```python
#visualisasi pertanyaan 3
account = day_df[['casual', 'registered']].sum().reset_index()
account.columns = ['User Type', 'Count']

plt.figure(figsize=(5, 3))
sns.barplot(
    x='User Type',
    y='Count',
    data=account,
    palette='viridis'
)

plt.title('Jumlah Pengguna Sepeda Terdaftar dan Tidak Terdaftar')
plt.xlabel('Tipe Pengguna')
plt.ylabel('Jumlah')
plt.show()
```

```python
#visualisasi pertanyaan 3
day_pattern = day_df.groupby('workingday').agg({
    'registered': 'sum',
    'casual': 'sum'
}).reset_index()

plt.figure(figsize=(5, 3))
plt.bar(
    day_pattern['workingday'],
    day_pattern['registered'],
    label='Registered',
    color='tab:red'
)
plt.bar(
    day_pattern['workingday'],
    day_pattern['casual'],
    bottom=day_pattern['registered'],
    label='Casual',
    color='tab:blue'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah Penyewa Sepeda Berdasarkan Hari Kerja dan Hari Libur')
plt.legend()
plt.show()
```

##Pertanyaan 4 : Apakah ada tren musiman dalam penggunaan sepeda?

```python
#visualisasi pertanyaan 4
plt.figure(figsize=(5,5))
sns.barplot(
    x='season',
    y='cnt',
    data=day_df,
    palette='viridis')

plt.title('Jumlah Pengguna Sepeda berdasarkan musim')
plt.xlabel('')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

## Conclusion

###Conclution pertanyaan 1 : Apakah ada pengaruh jumlah permintaan sepeda antara weekday dan holyday?

Jumlah permintaan sepeda pada hari kerja/*working day* lebih tinggi dibandingkan dengan hari libur/*Holiday*. hal ini menunjukkan bahwa lebih banyak orang menggunakan sepeda untuk perjalanan kerja atau kegiatan rutin lainnya di hari kerja



###Conclution pertanyaan 2 :Bagaimana pengaruh cuaca terhadap permintaan sepeda?

- Jumlah pengguna sepeda sangat tinggi pada cuaca *clear, cloudy*. hal ini berarti pengguna sepeda akan berkurang bila cuaca mulai buruk atau lebih buruk lagi
- Terdapat korelasi positif antara suhu (temp, atemp) dan jumlah permintaan sepeda. Ini berarti bahwa ketika suhu meningkat, jumlah permintaan sepeda juga cenderung meningkat.
- korelasi negatif antara kelembaban (hum) dan jumlah permintaan sepeda, menunjukkan bahwa kondisi lembab mengurangi penggunaan sepeda.
- Kecepatan angin (windspeed) juga menunjukkan korelasi negatif dengan jumlah permintaan sepeda.

###Conclution Pertanyaan 3 : Bagaimana penggunaan sepeda berbeda antara pengguna terdaftar dan pengguna tidak terdaftar?

Pengguna terdaftar cenderung menggunakan sepeda lebih sering dibandingkan pengguna tidak terdaftar. Rata-rata jumlah penggunaan sepeda oleh pengguna terdaftar lebih tinggi. Ini menunjukkan bahwa pengguna terdaftar mungkin lebih sering menggunakan sepeda untuk keperluan rutin seperti perjalanan ke tempat kerja, sementara pengguna tidak terdaftar lebih cenderung menggunakan sepeda untuk rekreasi.

##Conclution Pertanyaan 4 : Apakah ada tren musiman dalam penggunaan sepeda?

Penggunaan sepeda memuncak pada musim gugur. Musim gugur sering kali memiliki suhu yang lebih nyaman dibandingkan musim panas yang mungkin terlalu panas dan musim dingin yang mungkin terlalu dingin. Suhu yang sejuk membuat aktivitas bersepeda lebih menyenangkan
